```python
#use SHAP to explain decisions made by models

import shap
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC

#load the data
data = load_breast_cancer()
print(data)

#set the datasets
X, y = data['data'], data['target']
#malignant = 0 | benign = 1

#data['target_names']
```

```python
#split the data for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#scale the data for LinearSVC
clf = LinearSVC()
clf.fit(X_train, y_train)

#check accurancy
clf.score(X_test, y_test)
```

```python
#explain results for LinearSVC
#pass the prediction data and the train data
explainer = shap.KernelExplainer(clf.predict, shap.kmeans(X_train, 10))

shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values, X_test, feature_names=data['feature_names'])
```

```python
shap.initjs()
#shows forces applied in terms for the model to make a decision - using certain data
shap.force_plot(explainer.expected_value, shap_values[5], X_test[5], feature_names=data['feature_names']) 
```

```python
#aggreagate the values and take the abs value
mean_abs_shap_values = np.abs(shap_values).mean(axis=0)
feature_names = data['feature_names']
feature_importance = zip(feature_names, mean_abs_shap_values)

#sort by the most important feature that the model uses to make decisions
sorted_feature_importances = sorted(feature_importance, key=lambda x:x[1], reverse=True)

sorted_feature_importances
```

```python
#loop the set

for feature, importance in sorted_feature_importances:
    print(feature, " : ", importance)
```
